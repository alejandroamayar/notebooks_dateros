```hy
(import pandas)
(import numpy)
(import matplotlib)
(import seaborn)
```

```hy
%matplotlib notebook
```

```hy
(setv maga_fitosanitario (pandas.read_csv "MAGA - CERTIFICADOS FITOSANITARIOS - LIMPIO.csv"))
```

```hy
maga_fitosanitario
```

```hy
(.head (get maga_fitosanitario "Kg. Netos" ))
```

```hy
(-> maga_fitosanitario (.groupby "Categoría") (.sum) (.sort_values "CIF $" :ascending False) (.head 85))
```

```hy
(setv categorias_producto (-> maga_fitosanitario (.pivot_table
    :index '("Categoría" "Producto") 
    :values '("CIF $" "Permiso" "Kg. Netos")  
    :aggfunc {"CIF $" numpy.sum "Permiso" len "Kg. Netos" numpy.sum}
    :fill_value 0)
    (.sort_values "CIF $" :ascending False)
    (.head 85)
    ))
```

```hy
(seaborn.set :style "whitegrid")
(setv f (matplotlib.pyplot.subplots) ax (matplotlib.pyplot.subplots))
```

```hy
(matplotlib.pyplot.show (seaborn.stripplot :x "CIF $" :y "Kg. Netos" :hue "Permiso" :jitter True :data categorias_producto))
```
